```python
import matplotlib.pyplot as mplt
%matplotlib inline
```

```python
import  pandas as pd
```

```python
df = pd.read_csv("C:/Users/Askim/Downloads/data_2023-Dec-14.csv")
df
```

```python
Sorted_df = df.sort_values(by='date', ascending = True)
```

```python
x = Sorted_df['date']
```

```python
x
```

```python
y = Sorted_df['newCasesBySpecimenDate'].round(5)
y
```

```python
mplt.plot(x, y)
```

```python
import mplcursors
fig = mplt.figure(figsize=(18,8))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, y)
axes.set_title('Covid Cases in UK')
mplcursors.cursor(hover = True)
mplt.show()
```
